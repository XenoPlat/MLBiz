#### 1. Если мы рассчитали 95% доверительный интервал для среднего значения, то какие из следующих утверждений являются верными?

Выберите несколько вариантов и попытайтесь объяснить свой выбор:

1) Если многократно повторять эксперимент, то 95 % выборочных средних значений будут принадлежать рассчитанному нами доверительному интервалу.<br>
* _Не верно, т.к. это следствие из утверждения '2)'. Другими словами, доверительным интервалом мы оцениваем истинное среднее генеральной совокупности, а не возможные выборочные средние._

2) Мы можем быть на 95% уверены, что среднее значение в генеральной совокупности принадлежит рассчитанному доверительному интервалу.<br>
* _Верно, т.к. оно следует из свойств нормального распределения, когда с 95%-й надежностью можно быть уверенным, что оцениваемая величина будет принадлежать доверительному интервалу._<br> 

3) Если многократно повторять эксперимент, для каждой выборки рассчитывать свой доверительный интервал, то в 95 % случаев истинное среднее будет находиться внутри доверительного интервала.<br>

* _Верно, т.к. соответсвует определению ЦПТ_<br>

4) Среднее значение в генеральной совокупности точно превышает нижнюю границу 95% доверительного интервала.<br>
* _Не верно. Слово "точно" неуместно. Уместнее было бы "с доверительной вероятностью"_<br>

5) Среднее значение в генеральной совокупности точно принадлежит рассчитанному доверительному интервалу.<br>
* _Не верно. Слово "точно" неуместно. Уместнее было бы "с доверительной вероятностью"_

#### 2. Если мы увеличиваем объем выборки в два раза (при условии, что показатель стандартного отклонения остается неизменным), то 95% доверительный интервал

Выберите один вариант из списка

1) стал более узким<br>
2) возможны оба варианта<br>
3) стал более широким<br>

_Вариант __'1)'__: исходя из постулатов ЦТП, в формуле расчета доверительного интервала стандартное отклонение делится на корень из числа наблюдений:<br> X +- Zα/2*(σ/√n),<br>где X – математическое ожидание выборки, α – уровень статистической значимости, Zα/2 – Z-оценка для уровня статистической значимости α/2, σ – среднеквадратическое отклонение генеральной совокупности (стандартное отклонение), n – количество наблюдений в выборке. При этом, σ/√n является стандартной ошибкой.<br> Т.о., чем больше наблюдений, тем более узким становится доверительный интервал_

#### 4. Часто на практике нулевая гипотеза отклоняется, и различия считаются статистически достоверными, если p < 0,05. Однако часто в статистике используется более жесткий критерий достоверности различий, например, при условии, что p < 0,01. Значение p-уровня значимости, которое выбирается, в качестве порога обозначается буквой α (альфа). Например, если исследователь решил, что α = 0,05, то и нулевая гипотеза будет отклоняться при условии, что p < 0,05. 


#### Если в определенной ситуации весьма рискованно отклонить нулевую гипотезу, когда она на самом деле верна, то лучше использовать показатель α равный

1) 0,1<br>
2) 0,001<br>
3) 0,05<br>
4) 0,5<br>

_Вариант __'2)':__ ошибкой первого рода является ситуация, когда верная нулевая гипотеза отвергается; т.о., если есть риск отклонения верной нулевой гипотезы, то необходимо уменьшать уровень значимости α, т.е. вероятность получения ложно положительного результата._
